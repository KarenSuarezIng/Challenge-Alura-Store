### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head(10)
```

#1. Análisis de facturación



```python
ingresoTotalT1 = 0.0
ingresoTotalT2 = 0.0
ingresoTotalT3 = 0.0
ingresoTotalT4 = 0.0
```

```python
ingresos = [
    tienda['Precio'].sum(),
    tienda2['Precio'].sum(),
    tienda3['Precio'].sum(),
    tienda4['Precio'].sum()
]

for i in range(4):
    print(f"Ingreso total tienda {i+1}: {ingresos[i]}")
```

# 2. Ventas por categoría

```python
tiposCategoria = tienda['Categoría del Producto'].unique()
print(tiposCategoria)
```

```python
totalProductoCategoria = [
    tienda['Categoría del Producto'].value_counts(),
    tienda2['Categoría del Producto'].value_counts(),
    tienda3['Categoría del Producto'].value_counts(),
    tienda4['Categoría del Producto'].value_counts()
]

for i in range(len(totalProductoCategoria)):
  print(f"Ventas por categoría para la tienda {i+1}: ", totalProductoCategoria[i])
```

# 3. Calificación promedio de la tienda


```python
calificacionPromedio = [
    tienda['Calificación'].sum() / len(tienda['Calificación']),
    tienda2['Calificación'].sum() / len(tienda2['Calificación']),
    tienda3['Calificación'].sum() / len(tienda3['Calificación']),
    tienda4['Calificación'].sum() / len(tienda4['Calificación'])
]

for i in range(4):
  print(f"La calificación promedio de la tienda {i+1} es {round(calificacionPromedio[i],3)}")

```

# 4. Productos más y menos vendidos

```python
listaProductos = [
    tienda['Producto'].value_counts(),
    tienda2['Producto'].value_counts(),
    tienda3['Producto'].value_counts(),
    tienda4['Producto'].value_counts()
]

for i in range(4):
  print(f"Productos más vendidos para la tienda {i+1}: " , listaProductos[i].head(3), f"\Productos menos vendidos para la tienda {i+1}: ", listaProductos[i].tail(3), "\n\n")
```

# 5. Envío promedio por tienda

```python
envioPromedio = [
    tienda['Costo de envío'].sum() / len(tienda['Costo de envío']),
    tienda2['Costo de envío'].sum() / len(tienda2['Costo de envío']),
    tienda3['Costo de envío'].sum() / len(tienda3['Costo de envío']),
    tienda4['Costo de envío'].sum() / len(tienda4['Costo de envío'])
]

for i in range(4):
  print(f"El envío promedio de la tienda {i+1} es: {round(envioPromedio[i],3)}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Crear listas para almacenar top 3 de cada tienda
top_3_list = []

# Llenar las listas con los datos de productos más vendidos
for i in range(4):
    top = listaProductos[i].head(3).reset_index()
    top.columns = ['Producto', 'Cantidad']
    top['Tienda'] = f'Tienda {i+1}'
    top_3_list.append(top)

# Unir todos los top 3 en un solo DataFrame
top_df = pd.concat(top_3_list)

# Etiquetas de las tiendas
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# --- GRAFICAR ---
fig, axs = plt.subplots(1, 3, figsize=(24, 6))  # Ahora 3 columnas de gráficos

# 1. Gráfico de barras: Top 3 productos más vendidos
for tienda in top_df['Tienda'].unique():
    data = top_df[top_df['Tienda'] == tienda]
    axs[0].bar(data['Producto'] + " (" + tienda + ")", data['Cantidad'], label=tienda)

axs[0].set_title('Top 3 productos más vendidos por tienda')
axs[0].set_ylabel('Cantidad vendida')
axs[0].tick_params(axis='x', rotation=45)
axs[0].legend()

# 2. Gráfico de barras horizontales: costo de envío promedio
axs[1].barh(tiendas, envioPromedio, color='skyblue')
axs[1].set_title('Costo de envío promedio por tienda')
axs[1].set_xlabel('Costo promedio ($)')
axs[1].invert_yaxis()

# 3. Gráfico pastel: calificación promedio
axs[2].pie(calificacionPromedio, labels=tiendas, autopct='%1.1f%%', startangle=140)
axs[2].set_title('Distribución de calificación promedio por tienda')

# Ajustar espacio
plt.tight_layout()
plt.show()
```
